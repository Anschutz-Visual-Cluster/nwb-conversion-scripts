```python
# Create data structures for each dosage group
# animal_sessions = {
#     "pitx005": {"0.5mg/kg": [], "2mg/kg": [], "0mg/kg": []},
#     "pit006": {"0.5mg/kg": [], "2mg/kg": [], "0mg/kg": []},
#     "pit007": {"0.5mg/kg": [], "2mg/kg": [], "0mg/kg": []},
#     "pitx10": {"0.5mg/kg": [], "2mg/kg": [], "0mg/kg": []},
#     "pitx13": {"0.5mg/kg": [], "2mg/kg": [], "0mg/kg": []},
#     "pitx12": {"0.5mg/kg": [], "2mg/kg": [], "0mg/kg": []},
#     "pitx15": {"0.5mg/kg": [], "2mg/kg": [], "0mg/kg": []},
#     "pitx16": {"0.5mg/kg": [], "2mg/kg": [], "0mg/kg": []}
# }
animal_sessions = {}

# Sort and group sessions by animal_id and dosage
for session in crystalSessions:
    animal_id = session.nwb.subject.subject_id  # Get the animal_id
    sessionID = session.nwb.session_description  # Get the dosage

    #Print the extracted animal_id and dosage
    #print(f"Session: Animal ID: {animal_id}, Dosage: {sessionID}")

    # Map sessionID to a dosage key
    dosage = None
    if sessionID == "Dosage 0.5mg/kg":
        dosage = "0.5mg/kg"
    elif sessionID == "Dosage 2mg/kg":
        dosage = "2mg/kg"
    elif sessionID == "Dosage 0mg/kg":
        dosage = "0mg/kg"
# ( aid in animals) AND (dosage == true)  # a = "" -> false a = "asdjasflajs" -> true
    # Add the session to the corresponding animal_id and dosage group
    if animal_id not in animal_sessions:
        animal_sessions[animal_id] = {}
    if dosage not in animal_sessions[animal_id]:
        animal_sessions[animal_id][dosage] = []
    
    animal_sessions[animal_id][dosage].append(session)

# Debugging: Check the count of sessions grouped by animal and dosage
for animal_id, dosages in animal_sessions.items():
    for dosage, sessions in dosages.items():
        print(f"Animal {animal_id}, Dosage '{dosage}': {len(sessions)} sessions")

dosages = ["0mg/kg", "0.5mg/kg", "2mg/kg"]
directions = ["nasal", "temporal"]
print(animal_sessions)
# { "nasal": {
#              "0mg/kg":{ 
#                         "velocities": [..],  # (saccadenum,) a single number list [1.4, ..]
#                         "amps": [..], 
#                         "durations: [...},  # (saccadenum, start/stop) [[start, stop], [start, stop], ..] 
#                         "times: [...]
#               }, 
#               "0.5mg/kg": {{...}, ...},
#               <other dosages, etc>
#           }
# "temporal: {
#           <dosages>:<vels, amps, etc>, ...
#           }
# }

# amp_vels_durations_by_direction_and_dosage["nasal"]["0mg/kg"]["velocities"]
# amp_vels_durations_by_direction_and_dosage["nasal"]["0mg/kg"]["animal_name"]
print("="*50)
def calc_our_data():
    amp_vels_durations_by_direction_and_dosage = {}
    for animal_id in animal_sessions.keys():
        for dosage in dosages:
            print(f"Available dosages for animal '{animal_id}': '{list(animal_sessions[animal_id].keys())}'")
            animal_dosage_sessions = animal_sessions[animal_id][dosage]
            for sess in animal_dosage_sessions:
                for direc in directions:
                    all_waveforms = sess.pull(f"PredictSaccades.saccades_predicted_{direc}_waveforms")
                    epochs = sess.pull(f"PredictSaccades.saccades_predicted_{direc}_epochs")
                    peaks = sess.pull(f"PredictSaccades.saccades_predicted_{direc}_peak_indices")
                    # print(f"{animal_id}-{dosage}-{direc} wavs shape {waveforms.shape}")
                    # print(f"{animal_id}-{dosage}-{direc} epochs shape {epochs.shape}")
                    
                    amps = []
                    durations = []
                    vels = []
                    saccade_times = []
                    
                    for saccadenum in range(all_waveforms.shape[0]):
                        start_and_stop = (epochs[saccadenum] - peaks[:, None][saccadenum]) + (all_waveforms.shape[1]/2)  # time axis shape = 80 divide by 2 is 40 (since saccade is centered)
                        start = int(start_and_stop[0])
                        stop = int(start_and_stop[1])
                        waveforms = all_waveforms[saccadenum, start:stop]
                        
                        amp = np.max(waveforms) - np.min(waveforms)
                        amps.append(amp)
                        duration = (epochs[saccadenum][1] - epochs[saccadenum][0])/200  # Duration in frames
                        vels.append(amp / duration)
                        durations.append(duration)
                        saccade_times.append(epochs[saccadenum][0])
                    if direc not in amp_vels_durations_by_direction_and_dosage:
                        amp_vels_durations_by_direction_and_dosage[direc] = {}
                    if dosage not in amp_vels_durations_by_direction_and_dosage[direc]:
                        amp_vels_durations_by_direction_and_dosage[direc][dosage] = {}
                    
                    animal_names_list = [animal_id]*all_waveforms.shape[0]  # Create a repeated animal name list like ["samename", "samename", repeats with same data..]
                    # The repeated names are a bit redundant but useful to match a specific saccade to an animal name
                    for varname, vardata in [("velocity", vels), ("amplitude", amps), ("saccade_times", saccade_times), ("durations", durations), ("animal_name", animal_names_list)]:
                        # Extend takes two lists like [1,2,3] and [4,5,6] into [1,2,3,4,5,6]
                        # APPENDING DOES [1,2,3].append([4,5,6]) -> [1,2,3, [4,5,6] ]
                        if varname not in amp_vels_durations_by_direction_and_dosage[direc][dosage]:
                            amp_vels_durations_by_direction_and_dosage[direc][dosage][varname] = []
                        amp_vels_durations_by_direction_and_dosage[direc][dosage][varname].extend(vardata)
    
    return amp_vels_durations_by_direction_and_dosage

amp_vels_durations_by_direction_and_dosage = calc_our_data()

print("="*50)
for direc in ["nasal", "temporal"]:
    for dosage in ["0mg/kg", "0.5mg/kg", "2mg/kg"]:
        for varname in ["velocity", "amplitude", "saccade_times", "durations", "animal_name"]:
            amp_vels_durations_by_direction_and_dosage[direc][dosage][varname] = np.array(amp_vels_durations_by_direction_and_dosage[direc][dosage][varname])
            print(f"{direc}-{dosage}-{varname} {amp_vels_durations_by_direction_and_dosage[direc][dosage][varname].shape}")
print("="*50)


```

```python
#plot ipsilateral wt vs control d uration and  single points
import numpy as np
import matplotlib.pyplot as plt

# Define the dosages and corresponding titles
dosages = ["0mg/kg", "0.5mg/kg", "2mg/kg"]
titles = ['0mg/kg', '0.5mg/kg', '2mg/kg']

# Extract individual animal data (example: "pitx005" from DREADD and "pitx012" from Control)
single_het_animal = "pitx005"
single_wt_animal = "pitx012"

# Initialize lists to store individual durations for the selected animals
het_individual_durations = []
wt_individual_durations = []

# Loop through each dosage and extract the individual Het animal and WT animal durations
for dosage in dosages:
    animal_names_het = amp_vels_durations_by_direction_and_dosage["temporal"][dosage]["animal_name"]
    animal_names_wt = amp_vels_durations_by_direction_and_dosage["temporal"][dosage]["animal_name"]
    
    # Extract Het individual
    het_idx = animal_names_het == single_het_animal
    het_durations = amp_vels_durations_by_direction_and_dosage["temporal"][dosage]["durations"][het_idx]
    het_individual_durations.append(het_durations[0])  # Assuming a single entry for the individual
    
    # Extract WT individual
    wt_idx = animal_names_wt == single_wt_animal
    wt_durations = amp_vels_durations_by_direction_and_dosage["temporal"][dosage]["durations"][wt_idx]
    wt_individual_durations.append(wt_durations[0])  # Assuming a single entry for the individual

# Bar plot for means (already calculated means and stds)
x = np.arange(len(titles))  # the label locations
width = 0.9  # the width of the bars

fig, ax = plt.subplots(figsize=(10, 6))

# Adjust the x-axis so that all WT bars are on the left and all Het bars are on the right
bars_wt = ax.bar(x - 0.5*width, mean_temporal_durations_wt, width, yerr=std_temporal_durations_wt, label='Control', color='palevioletred', capsize=5)
bars_het = ax.bar(x + 4*width, mean_temporal_durations_het, width, yerr=std_temporal_durations_het, label='DREADD', color='rebeccapurple', capsize=5)

# Overlay individual animal points using scatter (shifted to the appropriate sides)
ax.scatter(x - 0.5*width, wt_individual_durations, color='darkmagenta', label=f'Control {single_wt_animal}', zorder=5, s=100)
ax.scatter(x + 4*width, het_individual_durations, color='midnightblue', label=f'DREADD {single_het_animal}', zorder=5, s=100)

# Connect individual Het animal data points with a line
ax.plot(x + 4*width, het_individual_durations, color='midnightblue', linestyle='-', linewidth=2, zorder=4)

# Connect individual WT animal data points with a line
ax.plot(x - 0.5*width, wt_individual_durations, color='darkmagenta', linestyle='-', linewidth=2, zorder=4)

# Add labels, title, and x-axis ticks
ax.set_ylabel("Duration (Seconds)", fontsize=20)
ax.set_title("Ipsilateral", fontsize=22)
ax.set_xticks(x)
ax.set_xticklabels(titles)
ax.legend(fontsize='x-large')
ax.tick_params(axis='y', labelsize=18)

# Set a reasonable y-limit if needed
plt.ylim(0,0.10)

plt.tight_layout()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the dosages and corresponding titles
dosages = ["0mg/kg", "0.5mg/kg", "2mg/kg"]
titles = ['0mg/kg', '0.5mg/kg', '2mg/kg']
het_animal_names = ["pitx005", "pitx013", "pitx014", "pitx015", "pitx016"]
wt_animal_names = ["pitx012", "pitx010"]

# Initialize lists to store means and standard deviations
mean_nasal_amplitude_het = []
std_nasal_amplitude_het = []
mean_nasal_amplitude_wt = []
std_nasal_amplitude_wt = []

# Initialize lists to store individual amplitudes for specific animals
single_het_amplitude = []
single_wt_amplitude = []

# Loop through each dosage to calculate means and standard deviations for Het animals
for dosage in dosages:
    animal_data = amp_vels_durations_by_direction_and_dosage["nasal"][dosage]
    animal_names = np.array(animal_data["animal_name"])
    animal_idxs_het = np.logical_or.reduce([animal_names == name for name in het_animal_names])
    
    # Extract amplitude for Het animals
    nasal_amplitude_het = animal_data["amplitude"][animal_idxs_het]
    
    # Calculate mean and standard deviation for Het
    mean_nasal_amplitude_het.append(np.mean(nasal_amplitude_het))
    std_nasal_amplitude_het.append(np.std(nasal_amplitude_het))

    # Store individual amplitude for the selected Het animal (pitx005)
    single_het_amplitude.append(nasal_amplitude_het[animal_names[animal_idxs_het] == "pitx005"].mean())  # Get average amplitude for pitx005

# Loop through each dosage to calculate means and standard deviations for WT animals
for dosage in dosages:
    animal_data = amp_vels_durations_by_direction_and_dosage["nasal"][dosage]
    animal_names = np.array(animal_data["animal_name"])
    animal_idxs_wt = np.logical_or.reduce([animal_names == name for name in wt_animal_names])
    
    # Extract amplitude for WT animals
    nasal_amplitude_wt = animal_data["amplitude"][animal_idxs_wt]
    
    # Calculate mean and standard deviation for WT
    mean_nasal_amplitude_wt.append(np.mean(nasal_amplitude_wt))
    std_nasal_amplitude_wt.append(np.std(nasal_amplitude_wt))

    # Store individual amplitude for the selected WT animal (pitx012)
    single_wt_amplitude.append(nasal_amplitude_wt[animal_names[animal_idxs_wt] == "pitx012"].mean())  # Get average amplitude for pitx012

# Create the bar plot with error bars
x = np.arange(len(titles))  # the label locations
width = 0.9  # the width of the bars

fig, ax = plt.subplots(figsize=(10, 6))

# Bar plots for Het and WT animals with error bars

bars_wt = ax.bar(x - 0.5*width, mean_nasal_amplitude_wt, yerr=std_nasal_amplitude_wt, label='Control', color='palevioletred', capsize=5)
bars_het = ax.bar(x + 4*width, mean_nasal_amplitude_het, yerr=std_nasal_amplitude_het, label='DREADD', color='rebeccapurple', capsize=5)

# Plot individual animal amplitudes as scatter points

ax.scatter(x - 0.5*width, single_wt_amplitude, color='darkmagenta', label='Control pitx012', zorder=5, s=100)
ax.scatter(x + 4*width, single_het_amplitude, color='midnightblue', label='DREADD pitx005', zorder=5, s=100)

# Connect the individual points with a line
ax.plot(x + 4*width, single_het_amplitude, color='midnightblue', marker='o', linestyle='-', linewidth=2, zorder=4)  # Line for Het animal
ax.plot(x - 0.5*width, single_wt_amplitude, color='darkmagenta', marker='o', linestyle='-', linewidth=2, zorder=4)  # Line for WT animal

# Add labels, title, and x-axis ticks
#ax.set_xlabel("Dosage")
ax.set_ylabel("Amplitude (Pixels)", fontsize=20)
ax.set_title("Contralateral", fontsize=22)
#ax.set_xticks(x)
#ax.set_xticklabels(titles)
ax.legend(fontsize='x-large')
ax.tick_params(axis='y', labelsize=18)

# Set y-limit based on your data
plt.ylim(0, 20)  # Adjust as necessary

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the dosages and corresponding titles
dosages = ["0mg/kg", "0.5mg/kg", "2mg/kg"]
titles = ['0mg/kg', '0.5mg/kg', '2mg/kg']
het_animal_names = ["pitx005", "pitx013", "pitx014", "pitx015", "pitx016"]
wt_animal_names = ["pitx012", "pitx010"]

# Initialize lists to store means and standard deviations
mean_temporal_amplitude_het = []
std_temporal_amplitude_het = []
mean_temporal_amplitude_wt = []
std_temporal_amplitude_wt = []

# Initialize lists to store individual amplitudes for specific animals
single_het_amplitude = []
single_wt_amplitude = []

# Loop through each dosage to calculate means and standard deviations for Het animals
for dosage in dosages:
    animal_data = amp_vels_durations_by_direction_and_dosage["temporal"][dosage]
    animal_names = np.array(animal_data["animal_name"])
    animal_idxs_het = np.logical_or.reduce([animal_names == name for name in het_animal_names])
    
    # Extract amplitude for Het animals
    temporal_amplitude_het = animal_data["amplitude"][animal_idxs_het]
    
    # Calculate mean and standard deviation for Het
    mean_temporal_amplitude_het.append(np.mean(temporal_amplitude_het))
    std_temporal_amplitude_het.append(np.std(temporal_amplitude_het))

    # Store individual amplitude for the selected Het animal (pitx005)
    single_het_amplitude.append(temporal_amplitude_het[animal_names[animal_idxs_het] == "pitx005"].mean())  # Get average amplitude for pitx005

# Loop through each dosage to calculate means and standard deviations for WT animals
for dosage in dosages:
    animal_data = amp_vels_durations_by_direction_and_dosage["temporal"][dosage]
    animal_names = np.array(animal_data["animal_name"])
    animal_idxs_wt = np.logical_or.reduce([animal_names == name for name in wt_animal_names])
    
    # Extract amplitude for WT animals
    temporal_amplitude_wt = animal_data["amplitude"][animal_idxs_wt]
    
    # Calculate mean and standard deviation for WT
    mean_temporal_amplitude_wt.append(np.mean(temporal_amplitude_wt))
    std_temporal_amplitude_wt.append(np.std(temporal_amplitude_wt))

    # Store individual amplitude for the selected WT animal (pitx012)
    single_wt_amplitude.append(temporal_amplitude_wt[animal_names[animal_idxs_wt] == "pitx012"].mean())  # Get average amplitude for pitx012

# Create the bar plot with error bars
x = np.arange(len(titles))  # the label locations
width = 0.9  # the width of the bars

fig, ax = plt.subplots(figsize=(10, 6))

# Bar plots for Het and WT animals with error bars

bars_wt = ax.bar(x - 0.5*width, mean_temporal_amplitude_wt, yerr=std_temporal_amplitude_wt, label='Control', color='palevioletred', capsize=5)
bars_het = ax.bar(x + 4*width, mean_temporal_amplitude_het, yerr=std_temporal_amplitude_het, label='DREADD', color='rebeccapurple', capsize=5)

# Plot individual animal amplitudes as scatter points

ax.scatter(x - 0.5*width, single_wt_amplitude, color='darkmagenta', label='Control pitx012', zorder=5, s=100)
ax.scatter(x + 4*width, single_het_amplitude, color='midnightblue', label='DREADD pitx005', zorder=5, s=100)

# Connect the individual points with a line
ax.plot(x + 4*width, single_het_amplitude, color='midnightblue', marker='o', linestyle='-', linewidth=2, zorder=4)  # Line for Het animal
ax.plot(x - 0.5*width, single_wt_amplitude, color='darkmagenta', marker='o', linestyle='-', linewidth=2, zorder=4)  # Line for WT animal

# Add labels, title, and x-axis ticks
#ax.set_xlabel("Dosage")
ax.set_ylabel("Amplitude (Pixels)", fontsize=20)
ax.set_title("Ipsilateral", fontsize=22)
#ax.set_xticks(x)
#ax.set_xticklabels(titles)
ax.legend(fontsize='x-large')
ax.tick_params(axis='y', labelsize=18)

# Set y-limit based on your data
plt.ylim(0, 20)  # Adjust as necessary

plt.tight_layout()
plt.show()
```
